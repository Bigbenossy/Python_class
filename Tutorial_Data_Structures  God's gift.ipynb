**Introduction to Tuple**

- A tuple is a collection data or values that can be of same data type or not
- Tuple is immutable. That is, once created, its values can not be changed
- Tuples are useful when you want to represent a collection of items that should not be modified.
- Examples of use cases: coordinates, database records, function returning multiple values.


```python
# Tuple Declaration
empty = ()
print(empty,type(empty))
```

```python
# Inserting values into tuple by separating each value with a comma
mytup= (2)
print(mytup)
```

```python
# assign a value to a tuple without the bracket
mytup = 2,3,4,5,5,6
print(mytup)
```

```python
# A tuple with different data type
mytup = ('Banana',3,'berry')
print(mytup)
```

```python
# Accessing items in a tuple using index
mytup = (2,3,4,5,6,7,8,9)
print(mytup[3])
```

```python
# Slicing can be used to access a range of elements:
print(mytup[1:4])
```

```python
# # Change the value of an item in our tuple
# mytup[4] = 10

# # This will raise a TypeError
# print(mytup)
```

```python
# Find the index of an element (first occurrence):
print(mytup.index(4))
```

```python
# Sizes of items in a tuple
size = len(mytup)
print(size)
```

```python
# Count the numbers of occurance of an item in a tuple
mytup = (2,3,4,5,6,7,8,9,6,7,9,9)
print(mytup.count(9))
```

```python
#Tuple Packing and Unpacking
# You can create a tuple without using parentheses. This is called "tuple packing."
packed_tuple = 1, 2, 3
print(packed_tuple)
```

```python
# You can also unpack a tuple into variables:
x, y, z = packed_tuple

print(x)
print(y)
print(z)
```

```python
# nested tuple
my_nested_tup = ((0,1),(1,2),(2,3),(3,4))
print(my_nested_tup)
```

**Subsets**

```python
var1 = {1, 2, 3, 4, 5}
var2 = {3, 4, 5}
```

```python
# The `issubset` method is used to check if one set is a subset of another set.
is_subset = var2.issubset(var1)
print(f"Is var2 a subset of var1? {is_subset}")
```

```python
# The `issuperset` method is used to check if one set is a superset of another set.
is_superset = var1.issuperset(var2)
print(f"Is var1 a superset of var2? {is_superset}")
```

**Set Union**

```python
# add a set a set using union
var1 = {'apple', 'orange', 'banana', 'cherry',4,'apple',}
var2 = {7, 8, 'watermelon', 'banana', 'lemon', 'grape', 'orange'}
set3 = var1.union(var2)

print(set3)
```

**Set Intersection**

```python
var1 = {'apple', 'orange', 'banana', 'cherry',4,'apple',}
var2 = {7, 8, 'watermelon', 'banana', 'lemon', 'grape', 'orange'}

set3 = var1.intersection(var2)
print(set3)
```

**Set Difference**

```python
var1 = {'apple', 'orange', 'banana', 'cherry',4,'apple',}
var2 = {7, 8, 'watermelon', 'banana', 'lemon', 'grape', 'orange'}

set3 = var1.difference(var2)
print(set3)
```

**Set Symmetric Difference**

```python
var1 = {'apple', 'orange', 'banana', 'cherry',4,'apple',}
var2 = {7, 8, 'watermelon', 'banana', 'lemon', 'grape', 'orange'}

set3 = var1.symmetric_difference(var2)
print(set3)
```
